# Librerias

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import optimize
from scipy.constants import h, k, c
from sigfig import round #https://www.delftstack.com/howto/python/round-to-significant-digits-python/
import warnings
warnings.filterwarnings('ignore') #https://gennadylaptev.medium.com/suppress-warnings-in-python-682abb3d533c
```

# Ley de Wien

```python
def f(nu):
    return 5*np.exp(-nu) + nu - 5 

eta = optimize.newton(f, 5)  #https://docs.scipy.org/doc/scipy/reference/generated/scipy.optimize.newton.html

f"La raíz es: {eta}"
```

```python
b = (h*c)/(k*eta)

f"b tiene un valor de: {b} m*K"
```

# Constante de Stefan-Boltzman

```python
sigma = (2*(np.pi**5)*(k**4))/(15*(h**3)*(c**2))

f"La constante de Stefan-Boltzmann es: {sigma} W/(m^2*K^4)"
```

# Ley de Planck

```python
def I(lamb,T):
    return (2*np.pi*h*c**2)/(lamb**5)*(1/(np.exp((h*c)/(lamb*k*T))-1))

def lamb_max(T):
    return b/T

lamb_1,lamb_2,lamb_3 = np.linspace(0.1e-5, 75e-5, 10000),np.linspace(0.1e-8, 300e-8, 10000),np.linspace(0.1e-9, 100e-9, 10000)
T_1,T_2,T_3 = 100,5800,120000
lamb_max_1,lamb_max_2,lamb_max_3 = lamb_max(T_1),lamb_max(T_2),lamb_max(T_3)
```

```python
fig,ax = plt.subplots()
ax.plot( lamb_1, I(lamb_1,T_1))
ax.scatter(lamb_max_1,I(lamb_max_1,T_1)) 
ax.annotate(r"$\lambda_{max}=$"+f"{round(lamb_max_1,3)} m", 
                 (lamb_max_1,I(lamb_max_1,T_1)), 
                 textcoords="offset points", 
                 xytext=(100,-10),
                 ha='center') #https://queirozf.com/entries/add-labels-and-text-to-matplotlib-plots-annotation-examples
ax.set_xlabel(r"$\lambda(m)$")
ax.set_ylabel(r"$I(\frac{W}{m^3})$")
ax.set_title(f"Intensidad por longitud de onda para un cuerpo negro a T={T_1}K") 
plt.show()
```

```python
fig,ax = plt.subplots()
ax.plot( lamb_2, I(lamb_2,T_2))
ax.scatter(lamb_max_2,I(lamb_max_2,T_2)) 
ax.annotate(r"$\lambda_{max}=$"+f"{round(lamb_max_2,3)} m", 
                 (lamb_max_2,I(lamb_max_2,T_2)), 
                 textcoords="offset points", 
                 xytext=(75,-10),
                 ha='center') #https://queirozf.com/entries/add-labels-and-text-to-matplotlib-plots-annotation-examples
ax.set_xlabel(r"$\lambda(m)$")
ax.set_ylabel(r"$I(\frac{W}{m^3})$")
ax.set_title(f"Intensidad por longitud de onda para un cuerpo negro a T={T_2}K") 
plt.show()
```

```python
fig,ax = plt.subplots()
ax.plot( lamb_3, I(lamb_3,T_3))
ax.scatter(lamb_max_3,I(lamb_max_3,T_3)) 
ax.annotate(r"$\lambda_{max}=$"+f"{round(lamb_max_3,3)} m", 
                 (lamb_max_3,I(lamb_max_3,T_3)), 
                 textcoords="offset points", 
                 xytext=(75,-10),
                 ha='center') #https://queirozf.com/entries/add-labels-and-text-to-matplotlib-plots-annotation-examples
ax.set_xlabel(r"$\lambda(m)$")
ax.set_ylabel(r"$I(\frac{W}{m^3})$")
ax.set_title(f"Intensidad por longitud de onda para un Cuerpo negro a T={T_3}K") 
plt.show()
```
